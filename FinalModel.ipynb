```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
# dataset
mitbih_train_data = pd.read_csv('/content/gdrive/My Drive/BTP - 7 Sem/Database/mitbih_train.csv', header=None)
mitbih_test_data = pd.read_csv('/content/gdrive/My Drive/BTP - 7 Sem/Database/mitbih_test.csv', header=None)
```

```python
#length of training and testing dataset
print(len(mitbih_train_data))
print(len(mitbih_test_data))
```

```python
mitbih_train_data[187] = mitbih_train_data[187].astype(int)
# categories count of ( N,O,V,S,F)
categories_count = mitbih_train_data[187].value_counts()
print(categories_count)
```

```python
train_df = mitbih_train_data
train_df.shape
```

```python
# Figure to show that the dataset is imbalanced
# Almost 80% of the values belongs to Normal beats

plt.figure(figsize=(12,12))
plt.pie(mitbih_train_data[187].value_counts(), labels=['N','Q','V','S','F'], colors=['orange','green','blue','yellow','lightgreen'], autopct='%1.1f%%')
plt.gcf().gca().add_artist(plt.Circle((0,0),0.7,color='white'))
plt.title('Imbalanced beats dataset')
plt.show()
```

```python
# Use of sklearn.utils.resample to convert the imbalance dataset to balance one (20% each)
from sklearn.utils import resample

# Separating the dataset category wise
data_0 = train_df[train_df[187]==0]
data_1 = train_df[train_df[187]==1]
data_2 = train_df[train_df[187]==2]
data_3 = train_df[train_df[187]==3]
data_4 = train_df[train_df[187]==4]

# Downsampling the data_0 which contains Normal beats as it contains almost 80% of data
data_0_downsample = resample(data_0,replace=True,n_samples=20000,random_state=122)
# Upsampling all other categories so that their percent changes to 20% each
data_1_upsample   = resample(data_1,replace=True,n_samples=20000,random_state=123)
data_2_upsample   = resample(data_2,replace=True,n_samples=20000,random_state=124)
data_3_upsample   = resample(data_3,replace=True,n_samples=20000,random_state=125)
data_4_upsample   = resample(data_4,replace=True,n_samples=20000,random_state=126)

# Concating the resampled data again to training dataset
train_df = pd.concat([data_0_downsample,data_1_upsample,data_2_upsample,data_3_upsample,data_4_upsample])
```

```python
# Balanced dataset

plt.figure(figsize=(12,12))
plt.pie(train_df[187].value_counts(), labels=['N','Q','V','S','F'], colors=['orange','green','blue','yellow','lightgreen'], autopct='%1.1f%%')
plt.gcf().gca().add_artist(plt.Circle((0,0),0.7,color='white'))
plt.title('Balanced beats dataset after resampling')
plt.show()
```

```python
# Deep Neural Network Model (Transfer learning)
from tensorflow.keras.models import Model
from keras.layers import Input, Conv1D, MaxPool1D, Activation, Add, Dense, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import LearningRateScheduler

def repeated_model(x):
  # conv and ReLu
  conv2 = Conv1D(filters=32, kernel_size=5, activation='relu', padding='same')(x)
  # conv
  conv3 = Conv1D(filters=32, kernel_size=5, padding='same')(conv2)
  # Addition
  add1 = Add()([x, conv3])
  # ReLu
  relu2 = Activation(activation='relu')(add1)
  # Max_Pooling
  pool1 = MaxPool1D(pool_size=5, strides=2)(relu2)
  return pool1

def training_model(num_classes):
  #input layer
  input = Input(shape=(187,1))
  num_class = num_classes
  # conv layer
  conv1 = Conv1D(filters=32, kernel_size=5)(input)
  # *5 
  pool1 = repeated_model(conv1)
  pool2 = repeated_model(pool1)
  pool3 = repeated_model(pool2)
  pool4 = repeated_model(pool3)
  pool5 = repeated_model(pool4)
  # flatten
  flat1 = Flatten()(pool5)
  dense1 = Dense(32)(flat1)
  relu1 = Activation(activation='relu')(dense1)
  dense2 = Dense(32)(relu1)
  final_layer_size = 5
  dense3 = Dense(final_layer_size)(dense2)
  # Softmax
  output = Activation(activation='softmax')(dense3)
  return Model(inputs=input, outputs=output) 
```

```python
import tensorflow as tf
learning_schedule = tf.keras.optimizers.schedules.ExponentialDecay(0.001, decay_steps=10000, decay_rate=0.75)
adam = Adam(learning_rate=learning_schedule, beta_1=0.9, beta_2=0.999, amsgrad=False)
```

```python
M_mitbih = train_df.values
# X contains the beats
X_mitbih = M_mitbih[:,:-1]
# y contains the categories
y_mitbih = M_mitbih[:,-1]
num_classes = len(np.unique(y_mitbih))
print(num_classes)
# model training
model_mitbih = training_model(num_classes)
model_mitbih.summary()
```

```python
test_df = mitbih_test_data.values

# Separating X and y values for test dataset
X_test = test_df[:,:-1]
y_test = test_df[:,-1]

# Compiling the model using adam optimizer and sparse categorical crossentropy
model_mitbih.compile(optimizer=adam, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Model fitting 
# Epochs = 45
# batch size = 256
# Validation % = 15%
history = model_mitbih.fit(np.expand_dims(X_mitbih, axis=2), 
                          y_mitbih, 
                          validation_split=0.15,
                          epochs=45,
                          batch_size=256,
                          verbose=0)
```

```python
print(history.history.keys())
```

```python
# Plotting Accuracy
def plot_accuracy(history):
  plt.plot(history.history['accuracy'])
  plt.plot(history.history['val_accuracy'])
  plt.title('Model accuracy')
  plt.ylabel('accuracy')
  plt.xlabel('epoch')
  plt.legend(['accuracy', 'val_accuracy'], loc='upper left')
  plt.show()

# Plotting Loss
def plot_loss(history):
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.legend(['loss', 'val_loss'], loc='upper left')
  plt.show() 
```

```python
plot_accuracy(history)
plot_loss(history)
```

```python
# Accuracy 
results = model_mitbih.evaluate(np.expand_dims(X_test, axis=2), y_test, batch_size=128)
print(f"The accuracy on the testing set is {np.round(results[1]*100,1)}%")
```

```python
# Pediction /testing data shape
predictions = model_mitbih.predict(np.expand_dims(X_test, axis=2))
y_pred = np.argmax(predictions, axis=1)
print(y_pred.shape)
print(y_test.shape)
```

```python
# Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, y_pred)
print(cf_matrix)
```
